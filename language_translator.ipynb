```python
%%writefile app.py
import streamlit as st
from deep_translator import GoogleTranslator
language_mapping = {
    "English": "en",
    "Spanish": "es",
    "French": "fr",
    "German": "de",
    "Chinese": "zh-CN",
    "Japanese": "ja",
    "Arabic": "ar",
    "Hindi": "hi",
    "Russian": "ru",
    "Bengali": "bn"
}
st.title("Smart Language Translator")
text_to_translater=st.text_area("Enter the text to translate:")
selected_language_name=st.selectbox("Select target language:",list(language_mapping.keys()))
if st.button("Translate"):
  if text_to_translater.strip()!="":
    target_language=language_mapping[selected_language_name]
    translater = GoogleTranslator(source="auto",target=target_language)
    translated_text=translater.translate(text_to_translater)
    st.success("Translated Text:")
    st.write(translated_text)
  else:
    st.warning("Please enter text")
```
