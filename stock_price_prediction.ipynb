```python
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense,Dropout
```

```python
#Download Apple stock data
data=yf.download("AAPL",start="2015-01-01",end="2026-02-20")
```

```python
data.head()
```

```python
data = data[['Close']]
```

```python
data.shape
```

```python
data.head()
```

```python
data.tail()
```

```python
scalar = MinMaxScaler(feature_range=(0,1))
scaled_data = scalar.fit_transform(data)
```

```python
scaled_data
```

```python
def create_dataset(dataset,time_step=60):
   X,y=[],[]
   for i in range(time_step,len(dataset)):
       X.append(dataset[i-time_step:i,0])
       y.append(dataset[i-0])
   return np.array(X),np.array(y)

time_step =60
X,y = create_dataset(scaled_data,time_step)
```

```python
X= X.reshape(X.shape[0],X.shape[1],1)

print("X shape:",X.shape)
print("y shape:",y.shape)
```

```python
train_size = int(len(X)*0.8)

X_train = X[:train_size]
X_test = X[train_size:]

y_train = y[:train_size]
y_test = y[train_size:]
```

```python
model = Sequential()

model.add(SimpleRNN(units=50,return_sequences=True,input_shape=(X_train.shape[1],1)))
model.add(Dropout(0.2))

model.add(SimpleRNN(units=50))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mean_squared_error')

model.summary() #optional
```

```python
X_train.shape[1]
```

```python
history = model.fit(
    X_train,y_train,
    epochs=20,
    batch_size=32,
    validation_data=(X_test,y_test)
)
```

```python
predictions = model.predict(X_test)

#Inverse transform to get real price

predictions = scalar.inverse_transform(predictions.reshape(-1,1))
y_test_actual = scalar.inverse_transform(y_test.reshape(-1,1))
```

```python
from sklearn.metrics import mean_absolute_error

mae =  mean_absolute_error(y_test_actual,predictions)
print("MAE:",mae)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test_actual,label="Actual Price")
plt.plot(predictions,label="Predicted Price")
plt.title("Stock Price Prediction using Vanilla RNN")
plt.xlabel("Time")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
#Get last 60 days from full dataset
last_60_days = scaled_data[-60:]

#Reshape to match RNN input shape
last_60_days = last_60_days.reshape(1,60,1)
```

```python
#PREDICT NEXT DAY

next_day_scaled = model.predict(last_60_days)

#convert back to original_price
next_day_price = scalar.inverse_transform(next_day_scaled)
print("Predicted Next Day Closing Price:",next_day_price[0][0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()

model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
model.add(Dropout(0.2))

model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=30,
    batch_size=32,
    validation_data=(X_test, y_test)
)
```

```python
predictions = model.predict(X_test)
predictions = scalar.inverse_transform(predictions.reshape(-1, 1))
y_test_actual = scalar.inverse_transform(y_test.reshape(-1, 1))
```

```python
mae = mean_absolute_error(y_test_actual, predictions)
print("MAE:", mae)
```

```python
plt.figure(figsize=(12,6))
plt.plot(y_test_actual, label="Actual Price")
plt.plot(predictions, label="Predicted Price")
plt.title("Stock Price Prediction using LSTM")
plt.xlabel("Time")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```
