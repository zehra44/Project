```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = {
'title': [
'Deep Learning for Healthcare',
'Machine Learning in Finance',
'Natural Language Processing Basics',
'AI for Medical Diagnosis',
'Deep Learning for Image Recognition'
],

'abstract': [
'Deep learning techniques used in healthcare data',
'Machine learning models applied in financial prediction',
'NLP techniques for text processing',
'Artificial intelligence used for medical diagnosis',
'Deep learning used for image classification'
]
}

df = pd.DataFrame(data)
```

```python
df['content'] = df['title'] + " " + df['abstract']
```

```python
vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(df['content'])
```

```python
similarity = cosine_similarity(tfidf_matrix)
```

```python
def recommend_papers(title):

    index = df[df['title'] == title].index[0]

    similarity_scores = list(enumerate(similarity[index]))

    similarity_scores = sorted(similarity_scores,
                               key=lambda x: x[1],
                               reverse=True)

    top_papers = similarity_scores[1:4]

    for paper in top_papers:
        print(df.iloc[paper[0]]['title'])
```

```python
recommend_papers("Deep Learning for Healthcare")
```
